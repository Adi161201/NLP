```python
import nltk
import spacy
import warnings
warnings.filterwarnings('ignore')
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
words= ["eating","eats","eat","ate","adjustable","rafting","ability","meeting","better"]

for word in words:
    print(word, " | ", stemmer.stem(word))
```

```python
nlp=spacy.load('en_core_web_sm')

doc=nlp("eating eat ate adjustable rafting ability meeting better")
for token in doc:
    print(token , " | ", token.lemma_)
```

```python
nlp.pipe_names
```

```python
doc=nlp("Bro, you wanna go? brah, don't say no! I am exhausted")
for token in doc:
    print(token , " | ", token.lemma_)
```

We know that Bro and Brah have same meaning but NLP doesnt understand it , so we will define a rule so that it will lemmmitize them to same word.

```python
ar=nlp.get_pipe('attribute_ruler')
ar.add([ [{"TEXT":"Bro"}] , [{"TEXT":"brah"}] ], {"LEMMA":"Brother"})
doc=nlp("Bro, you wanna go? brah, don't say no! I am exhausted")
for token in doc:
    print(token , " | ", token.lemma_)
```
